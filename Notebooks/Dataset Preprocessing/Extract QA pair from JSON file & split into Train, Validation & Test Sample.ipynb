```python
import json
import random
import os

# Configurable split percentages (ensure they sum to 100)
split_percentages = {
    "train": 70,
    "validation": 15,
    "test": 15
}

# Input file path and format
input_path = "/content/drive/MyDrive/Master Thesis/Datasets/Fundamentals_of_Risk_&_Insurance/Fundamentals_of_Risk_and_Insurance_QA_Dataset_extracted.jsonl"  # Input file (either .json or .jsonl)
file_format = "jsonl"  # Change to "json" if using a JSON array instead

# Output directory path
#output_dir = "output_splits"  # You can change this to any desired path
output_dir = "/content/drive/MyDrive/Master Thesis/Datasets/Fundamentals_of_Risk_&_Insurance/output_splits"  # You can change this to any desired path

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)


# Load data
if file_format == "json":
    with open(input_path, "r", encoding="utf-8") as f:
        data = json.load(f)
else:
    with open(input_path, "r", encoding="utf-8") as f:
        data = [json.loads(line) for line in f]

# Shuffle data for random split
random.shuffle(data)

# Calculate split sizes
total = len(data)
train_end = int(split_percentages["train"] / 100 * total)
val_end = train_end + int(split_percentages["validation"] / 100 * total)

train_data = data[:train_end]
val_data = data[train_end:val_end]
test_data = data[val_end:]

# Write JSONL files to the specified directory
def write_jsonl(filename, dataset):
    path = os.path.join(output_dir, filename)
    with open(path, "w", encoding="utf-8") as f:
        for item in dataset:
            f.write(json.dumps(item) + "\n")

write_jsonl("train.jsonl", train_data)
write_jsonl("validation.jsonl", val_data)
write_jsonl("test.jsonl", test_data)

print(f"✅ Dataset split complete:")
print(f" - Train: {len(train_data)} samples → {os.path.join(output_dir, 'train.jsonl')}")
print(f" - Validation: {len(val_data)} samples → {os.path.join(output_dir, 'validation.jsonl')}")
print(f" - Test: {len(test_data)} samples → {os.path.join(output_dir, 'test.jsonl')}")
```
